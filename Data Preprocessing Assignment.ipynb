```python
import pandas as pd

def readFile(fileName, summary = False):
#Function reads GCT file into a Pandas DataFrame

    df = pd.read_csv(fileName, sep='\t',skiprows=2) #read in file 
    if (summary): #if summary parameter is passed, dimensions are printed.
        print("(rows, columns):", df.shape)
    return df
```

```python
readFile("BRCA_minimal_60x19.gct", True) 
#Example of the Function in action
```

```python
df = readFile("BRCA_large_20783x40.gct") #read in GCT file into DataFrame
```

```python
df_sum = pd.DataFrame([]) #df_sum represents DataFrame Summary
df_sum.insert(loc=0, column="Name", value=df["Name"]) #Sample name
df_sum.insert(loc=1, column="Mean", value=df.mean(axis=1)) #Mean
df_sum.insert(loc=2, column="Median", value=df.median(axis=1)) #Median
df_sum.insert(loc=3, column="Standard deviation", value=df.std(axis=1)) #Standard Deviation
```

```python
#This DataFrame computes the Mean, Median, and Standard Deviation of each sample in the GCT File.
df_sum
```

```python
df_filtered = df.iloc[:,2:][df.iloc[:,2:] >= 500] #filter rows with values less than 500.
df_filtered = df_filtered.dropna() #remove NaN values that was produced from first filtering.

#df_filtered represents the original df but without the samples whose maximum value is less than 500.
```

```python
df_filtered
```

```python
dff_sum = pd.DataFrame([]) #dff_sum represents DataFrame Filtered Summary
dff_sum.insert(loc=0, column="Mean", value=df_filtered.mean(axis=0)) #Compute filtered mean
dff_sum.insert(loc=1, column="Median", value=df_filtered.median(axis=0)) #Compute filtered Median
```
